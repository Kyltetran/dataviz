```python
music_characteristics = spotify[['bpm', 'mode', 'danceability_%', 'valence_%', 'energy_%']]
music_characteristics
```

```python
def assign_genre(bpm):
    genres = [
        {"name": "Ambient", "max": 60},
        {"name": "Ballad, Chillout", "max": 80},
        {"name": "R&B and Soul", "max": 90},
        {"name": "Hip Hop", "max": 115},
        {"name": "Techno", "max": 125},
        {"name": "House", "max": 130},
        {"name": "Electro", "max": 128},
        {"name": "Dubstep", "max": 145},
        {"name": "Drum and Bass", "max": 187}
    ]
    
    for genre in genres:
        if bpm <= genre["max"]:
            return genre["name"]
    return 'Other'

music_characteristics['GENRE'] = music_characteristics['bpm'].apply(assign_genre)
music_characteristics[['bpm', 'GENRE']]
```

```python
music_characteristics['GENRE'] = music_characteristics['bpm'].apply(assign_genre)

# Then, you set the genre column as the new index:
music_characteristics.set_index('GENRE', inplace=True)
music_characteristics.sort_index(inplace=True)
music_characteristics
```

```python
genre_counts = music_characteristics.groupby(level=0).size()
total_tracks = len(music_characteristics)
genre_percentage = (genre_counts / total_tracks) * 100
genre_percentage
```

```python
colors = sns.color_palette("tab20c")

plt.figure(figsize=(10, 7))
genre_percentage.plot.pie(autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Music Genre (%)', fontweight='bold')
plt.ylabel('') 
plt.show()
```

```python
mean_values = music_characteristics.groupby('GENRE').agg({
    'energy_%': 'mean',
    'danceability_%': 'mean',
    'valence_%': 'mean'
}).sort_values(by='energy_%')

mean_values
```

```python
music_characteristics[['danceability_%', 'valence_%', 'energy_%']].mean()
```

```python
music_characteristics[['danceability_%', 'valence_%', 'energy_%']].median()
```

```python
music_characteristics['mode'].value_counts()
```

```python
# Count the number of major and minor songs for each genre
major_minor_counts = music_characteristics.groupby(['GENRE', 'mode']).size().unstack(fill_value=0)

# Compute the percentage of major songs
major_minor_counts['major_%'] = (major_minor_counts['Major'] / (major_minor_counts['Major'] + major_minor_counts['Minor'])) * 100

# Now, merge this major percentage with your mean_values dataframe
music_characteristics_summary = mean_values.merge(major_minor_counts[['major_%']], left_index=True, right_index=True)

music_characteristics_summary
```

```python
bar_width = 0.2
colors = sns.color_palette("tab20c", 20)[::3]
music_characteristics_summary_sorted = music_characteristics_summary.sort_index()
r = np.arange(len(music_characteristics_summary))

plt.figure(figsize=(15,7))

for index, (key, values) in enumerate(music_characteristics_summary_sorted.items()):
    plt.bar([x + bar_width*index for x in r], values, width=bar_width, edgecolor='grey', label=key, color=colors[index])

plt.xlabel('Genres', fontweight='bold', fontsize=15)
plt.xticks([r + bar_width*1.5 for r in range(len(music_characteristics_summary_sorted))], music_characteristics_summary_sorted.index)
plt.legend()

plt.tight_layout()
plt.title('Music Characteristics (%)', fontweight='bold', fontsize=16)
plt.ylabel('%', fontweight='bold', fontsize=15)

plt.axhline(y=50, color='gray', linestyle='--', linewidth=0.8)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

plt.show()
```

```python
music_platform = spotify[['bpm', 'in_spotify_playlists', 'in_apple_playlists', 'in_deezer_playlists']]
music_platform
```

```python
def assign_genre(bpm):
    genres = [
        {"name": "Ambient", "max": 60},
        {"name": "Ballad, Chillout", "max": 80},
        {"name": "R&B and Soul", "max": 90},
        {"name": "Hip Hop", "max": 115},
        {"name": "Techno", "max": 125},
        {"name": "House", "max": 130},
        {"name": "Electro", "max": 128},
        {"name": "Dubstep", "max": 145},
        {"name": "Drum and Bass", "max": 187}
    ]
    
    for genre in genres:
        if bpm <= genre["max"]:
            return genre["name"]
    return 'Other'

music_platform['GENRE'] = music_platform['bpm'].apply(assign_genre)
music_platform[['bpm', 'GENRE']]
```

```python
music_platform['GENRE'] = music_platform['bpm'].apply(assign_genre)
music_platform = music_platform[['GENRE', 'in_spotify_playlists', 'in_apple_playlists', 'in_deezer_playlists']]
music_platform.reset_index(drop=True, inplace=True)
music_platform
```

```python
music_platform.sort_values(by="in_spotify_playlists", ascending=False)
```

```python
colors = ['#d2e6f4', '#89bed2', '#4f97c1', '#2e6791', '#184c6e', '#0b314d']
cmap = mcolors.LinearSegmentedColormap.from_list('custom', colors, N=6)

music_platform.rename(columns={'in_spotify_playlists': 'Spotify',
                               'in_apple_playlists': 'Apple',
                               'in_deezer_playlists': 'Deezer'}, inplace=True)

# Convert the desired columns to numeric
cols_to_convert = ['Spotify', 'Apple', 'Deezer']

for col in cols_to_convert:
    music_platform[col] = pd.to_numeric(music_platform[col], errors='coerce')


genre_grouped = music_platform.groupby('GENRE').mean()

# Create a heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(genre_grouped, annot=True, fmt=".1f", cmap=cmap, linewidths=.5)

plt.title('Average Song Popularity by Genre Across Platforms', fontweight='bold')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
artist = spotify[['artist(s)_name', 'streams']]
artist
```

```python
artist = pd.DataFrame(artist)

# Group by artist(s)_name, summing the streams and counting the number of tracks
artist = artist.groupby('artist(s)_name').agg(total_streams=('streams', 'sum'), track_count=('streams', 'size')).sort_values(by='total_streams', ascending=False)

artist
```

```python
top_artist = artist['track_count'].sort_values(ascending=False).head(20)
top_artist
```

```python
top_artist.plot(kind='bar', figsize=(10,6))
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.title('Number of Releases by Artist', fontweight='bold')
plt.ylabel('Number of Releases', fontweight='bold')
plt.xlabel(' ')
plt.show()
```

```python
head_artist = artist['total_streams'].sort_values(ascending=False).head(20)
head_artist
```

```python
ax = head_artist.plot(kind='bar', figsize=(10,6))
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.title('Total Streams by Artist', fontweight='bold')
plt.xlabel(' ', fontweight='bold')

# Adjusting the y-axis labels to represent in billions
y_labels = ['{:.1f}B'.format(y/(10**10)) for y in ax.get_yticks()]
ax.set_yticklabels(y_labels)

plt.ylabel('Total Streams', fontweight='bold')
plt.show()
```
