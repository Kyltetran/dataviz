```python
import pandas as pd
```

```python
df = pd.read_csv('heart-disease.csv')
```

```python
columns = df.columns.tolist()    
columns.remove('target')   
feature_data = df[columns]
target_data = df.target
```

```python
from sklearn.model_selection import train_test_split
trainX,testX, trainY,testY = train_test_split(feature_data, target_data, train_size=0.70)
print('Training:' + str(trainX.shape)) 
print('Test:' + str(testX.shape))  
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
model = Pipeline([('scaler', StandardScaler()),
                  ('linear_regression', LogisticRegression())])
model.fit(trainX, trainY)
```

```python
linear_model = model.named_steps['linear_regression']
print("Model intercept:", linear_model.intercept_)
print("Model coefficients:", linear_model.coef_)
```

```python
model.score(testX, testY)
```

```python
predictions = model.predict(testX)                # Predict labels for testX
```

```python
comparison = pd.DataFrame({
    'Sample Index': testY.index,
    'Actual Heart Disease': testY.values,
    'Predicted Heart Disease': predictions
})

comparison.head(20)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, classification_report

print("Precision: ", precision_score(model.predict(testX), testY))
print("Recall: ", recall_score(model.predict(testX), testY))
print("F1: ", f1_score(model.predict(testX), testY))
print("Confusion matrix: \n", confusion_matrix(model.predict(testX), testY))
print("Classification report: \n", classification_report(model.predict(testX), testY))
```
